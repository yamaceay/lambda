* Examine Lemma 1.6.5 and see if you can understand its proof, even if it is described as "not hard, but rather boring."

> Lm 1.6.5: Let $x \not\equiv y$ and assume $x \notin FV(L)$, then $M[x := N][y := L] \equiv M[y := L][x := N[y := L]]$

* * $M[x := N][y := L]$ would contain free x's if $x \in FV(L)$ and $y \in M[x := N]$. 
* * But this is not the case, since $x \notin FV(L)$. So, all $x$'s and $y$'s are replaced.
* * Now, consider $M[y := L]$: $y$'s are replaced by $L$'s.
* * If $y \in FV(N)$: $M[y := L][x := N]$ would yield new free $y$'s. But those $y$'s had to be replaced by $L$'s.
* * So a new statement $N' \equiv N[y := L]$ updates the $y$'s in $N$ to $L$'s. $M[y := L][x := N']$ would work.
* * Then: $M[x := N][y := L] \equiv M[y := L][x := N[y := L]]$

Reflect on the potential effects of dropping some of the compatibility rules in Definition 1.8.1.

* * Assume $M \equiv N$ but there is a $L$ s.t. $ML \not\equiv NL$ (or $LM \not\equiv LN$).
* * $M \equiv \lambda x : P$. Then $ML \to_\beta P[x := L]$ and $NL \to_\beta P[x := L]$. They cannot differ in their $\beta$-normal form.
* * $L \equiv \lambda x : P$. Then $LM \to_\beta P[x := M]$ and $LN \to_\beta P[x := N] \equiv P[x := M]$ have to be different. But this is not possible.
* * $\lambda z . M \equiv \lambda z . N$. This is already proven by the second point.
