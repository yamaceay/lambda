* Is there another inhabitant of the type $A \to B \to A$ on p 52?
    * There are infinitely many inhabitants of such terms. Let $K \equiv A \to B \to A$, then define $I : K \to K$ as identity function, so that $IK$ is another inhabitant, and $I(IK)$, and so on.
* Concerning PAT on p53: Some people use the slogan “proofs as programs”. Does this make sense to you?
    * Yes. An inhabitant proves that a type is legal, so it is the proof. Using the derivation steps, we can find the inhabitants, also the running program. Using the inference mechanisms to derive new statements, we can confirm / reject that a type is legal.

* The proof of Lemma 2.10.5 (3) states “Try this yourself.” Do so.
    * ${\cal J}$ has the form $\Gamma \vdash M : \sigma$
        * Permutation Lemma: Let $\Gamma'$ be a permutation of $\Gamma$. Treat (var) case as true, since the variable should already exist in $\Gamma'$, if it exists in $\Gamma$ as well. Now, in case of (appl): $\Gamma \vdash NL : \sigma$, we also know that there is a type $\rho$ s.t. $\Gamma \vdash N : \rho \to \sigma$ and $\Gamma \vdash L : \rho$. Assuming $N$ and $L$ are not an application, we can replace $\Gamma$ with $\Gamma'$, s.t. $\Gamma' \vdash N : \rho \to \sigma$ and $\Gamma' \vdash L : \rho$. This leads to $\Gamma' \vdash NL : \sigma$. Now, in case of (abst): $\Gamma \vdash \lambda x . P : \rho \to \tau$, it is equal to $\Gamma, x : \rho \vdash P : \tau$. Now, define $\Gamma_x := \Gamma \cup \{x : \rho\}$ and the same applies for $\Gamma'_x$. Assuming P is not an abstraction, $\Gamma_x \vdash P : \tau$ results as $\Gamma'_x \vdash P : \tau$ and $\Gamma' \vdash \lambda x . P : \rho \to \tau$ as well.

    * (var) $M \equiv x$, then $\text{sub}(M) = \{x\}$. If $M \equiv x$ is legal, all terms are legal.
    * (abst) $M \equiv \lambda x . P$, then $\text{sub}(M) = \{\lambda x . P\} \cup \text{sub}(P)$. We know that $M \equiv \lambda x . P$ is legal, so prove that $\text{sub}(P)$ is legal. 
    * (appl) $M \equiv NL$, then $\text{sub}(M) = \{NL\} \cup \text{sub}(N) \cup \text{sub}(L)$. We know that $M \equiv NL$ is legal, so prove that $\text{sub}(N)$ and $\text{sub}(L)$ are legal.
    * Recursion in (abst) and (appl) steps continue until the recursion anchoris reached, also the (var) level. Then, a variable is proven to be legal, and every call is satisfied in a bottom-up manner.
