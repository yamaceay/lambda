    * (form) 
        * $\Gamma \vdash A : s$ and $\Gamma \vdash B : s$ implies $\Gamma \vdash A \to B : s$
        * $\Gamma \vdash A : *$ and $\Gamma, x : A \vdash B : s$ implies $\Gamma \vdash \Pi x : A . B : s$
    * (appl) 
        * $\Gamma \vdash M : A \to B$ and $\Gamma \vdash N : A$ implies $\Gamma \vdash M N : B$
        * $\Gamma \vdash M : \Pi x : A . B$ and $\Gamma \vdash N : A$ implies $\Gamma \vdash M N : B[x := N]$
    * (abst) 
        * $\Gamma, x : A \vdash M : B$ and $\Gamma \vdash A \to B : s$ implies $\Gamma \vdash \lambda x : A . M : A \to B$
        * $\Gamma, x : A \vdash M : B$ and $\Gamma \vdash \Pi x : A . B : s$ implies $\Gamma \vdash \lambda x : A . M : \Pi x : A . B$

    * $\Pi$ reappears, since the terms $x$ may appear freely in type $A$.
    * $\lambda x : A . M$ is the only possible function format given $A : *$.
